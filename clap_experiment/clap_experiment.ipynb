```python
from transformers import AutoProcessor, ClapAudioModel

model = ClapAudioModel.from_pretrained("laion/clap-htsat-unfused")
processor = AutoProcessor.from_pretrained("laion/clap-htsat-unfused")
```

# ClapAudioModel

```python
import glob
import os

audio_folder = "/Users/iveshenry18/Documents/Composition/Ingredients (partial copy)"
# get all .wav paths in audio_folder
wav_paths = glob.glob(audio_folder + "/**/*.wav", recursive=True)
```

```python
from datasets import Dataset
from datasets import Audio

dataset = Dataset.from_dict({"audio": wav_paths }).cast_column("audio", Audio())
```

```python

# Load image at bongos_path for input into processor
inputs = processor(audios=[datum["audio"]["array"] for datum in dataset], return_tensors="pt", sampling_rate=48000)
```

```python

outputs = model(**inputs)
last_hidden_state = outputs.last_hidden_state
```

# ClapTextModel

```python
from transformers import AutoTokenizer, ClapTextModelWithProjection

model = ClapTextModelWithProjection.from_pretrained("laion/clap-htsat-unfused")
tokenizer = AutoTokenizer.from_pretrained("laion/clap-htsat-unfused")

inputs = tokenizer(["a sound of a cat", "a sound of a dog"], padding=True, return_tensors="pt")

outputs = model(**inputs)
text_embeds = outputs.text_embeds
```
